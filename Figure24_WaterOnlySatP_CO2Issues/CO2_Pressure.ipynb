```python
rho=2700
P_axis_min=0
P_axis_max=6
D_axis_min=100000*P_axis_min/(9.8*rho)
D_axis_max=100000*P_axis_max/(9.8*rho)

fig, (((ax6a),(ax6b)), ((ax1), (ax3))) = plt.subplots(2, 2, figsize=(11,5),gridspec_kw={'height_ratios': [1, 5]})
plt.subplots_adjust(wspace=0.35, hspace=0)
cut=0

ax6a.hist(Ras2['CO2'], ec='k', fc='None', bins=[0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000,
                           1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800], density=True)

ax6a.hist(cas['co2_glass_pec_ppm'], ec='k', fc='red', alpha=0.5, bins=[0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000,
                           1100, 1200, 1300, 1400, 1500, 1600, 1700, 1800], density=True)
ax6a.set_xlim([0, 1800])
ax1.set_xlim([0, 1800])

ax6b.hist(combo_bub, ec='k', fc='rosybrown', alpha=1, 
          bins=np.linspace(0, 5000, 21), density=True)
ax6b.set_xlim([0, 5000])
ax3.set_xlim([0, 5000])

ax1.plot(10000*df['CO2'], df['PSat_2wt']/1000, '-k')

ax1.plot(10000*df['CO2'], df['PSat_4wt']/1000, '-', color='grey')

ax1.plot(10000*df['CO2'], df['PSat_6wt']/1000, '-', color='cyan')

ax1.plot(10000*df['CO2'], df['PSat_8wt']/1000, '-', color='cornflowerblue')

ax1.plot(10000*df['CO2'], df['PSat_10wt']/1000, '-', color='blue')
#ax1.plot(10000*df['CO2'], df['PSat_10wt']/1000, '-', color='royalblue')


ax2 = ax1.twinx()
ax1.set_ylim([P_axis_max, P_axis_min])
# ax2.tick_params(axis='y',which='both',direction='out',right=True)
# ax2.tick_params(axis="y",direction="out", pad=-15)
ax1.set_yticks([1, 2, 3, 4,5 ,6, 7])


ax2.set_ylim([D_axis_max, D_axis_min])

################## The whole range of CO2 ##################

P_axis_min=0
P_axis_max=9
D_axis_min=100000*P_axis_min/(9.8*rho)
D_axis_max=100000*P_axis_max/(9.8*rho)

ax4 = ax3.twinx()
ax3.set_ylim([P_axis_max, P_axis_min])

ax3.plot(10000*df['CO2'], df['PSat_2wt']/1000, '-k')

ax3.plot(10000*df['CO2'], df['PSat_4wt']/1000, '-', color='grey')

ax3.plot(10000*df['CO2'], df['PSat_6wt']/1000, '-', color='cyan')

ax3.plot(10000*df['CO2'], df['PSat_8wt']/1000, '-', color='cornflowerblue')
ax4.set_ylim([D_axis_max, D_axis_min])


ax3.plot(10000*df['CO2'], df['PSat_10wt']/1000, '-', color='blue')
ax4.set_ylim([D_axis_max, D_axis_min])


ax6b.axes.get_yaxis().set_visible(False)
ax6b.axes.get_xaxis().set_visible(False)

ax6a.axes.get_yaxis().set_visible(False)
ax6a.axes.get_xaxis().set_visible(False)

ax1.grid(color = 'k', linestyle = '--', linewidth = 1, alpha = 0.1)
ax3.grid(color = 'k', linestyle = '--', linewidth = 1, alpha = 0.1)
ax3.plot([3700, 3700], [0, 9], ':r', zorder=0)

fig.savefig('CO2_vs_Press.png', dpi=200)
```
