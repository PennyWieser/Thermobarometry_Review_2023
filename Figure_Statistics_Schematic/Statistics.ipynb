```python

# High R2, but Gradient and Intercept wrong
y_highR2=np.random.rand(24)+x*0.65-1/5+4.65
# Low R2 but perfect match
y_nosys=np.random.rand(24)*10+x-5
# Small RMSE but large mean error
y_small_RMSE=np.random.rand(24)*5.3+x
y_small_MeanError=np.random.rand(24)*20+x-10
```

```python
Stats_highR2=pt.calculate_R2_np(x, y_highR2)
print(Stats_highR2['R2'])
print(Stats_highR2['RMSE'])
print(Stats_highR2['Grad'])
print(Stats_highR2['Int'])
print(Stats_highR2['Mean'])
```

```python
Stats_nosys=pt.calculate_R2_np(x, y_nosys)
print(Stats_nosys['R2'])
print(Stats_nosys['RMSE'])
print(Stats_nosys['Grad'])
print(Stats_nosys['Int'])
print(Stats_nosys['Mean'])
```

```python
Stats_smallRMSE=pt.calculate_R2_np(x, y_small_RMSE)
print(Stats_smallRMSE['R2'])
print(Stats_smallRMSE['RMSE'])
print(Stats_smallRMSE['Grad'])
print(Stats_smallRMSE['Int'])
print(Stats_smallRMSE['Mean'])
```
