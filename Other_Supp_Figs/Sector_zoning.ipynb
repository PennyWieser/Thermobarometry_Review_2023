```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
```

```python
out=pt.import_excel('Hammer_Ps.xlsx', sheet_name='Sheet1', suffix='_Cpx')
Cpx=out['Cpxs']
all=out['my_input']
```

```python
comps=pt.calculate_clinopyroxene_components(cpx_comps=Cpx)
```

In summary, according to this assessment, the best thermometer is
the model of Putirka et al. (2003), and the best barometer is
the clinopyroxene-only temperature-dependent barometer
of Putirka (2008) Eq. 32b, with input temperature calculated using the model of Putirka et al. (1996).

```python
P=pt.calculate_cpx_only_press_temp(cpx_comps=Cpx, equationT='T_Put2008_eq32d', equationP='P_Put2008_eq32a').P_kbar_calc
```

```python
spongy_Ka01=(all['texture']=='spongy')&(all['xstal']=='Ka01 crystals: Ka01-26 (mickey), Ka01-18 (Gizmo)')
nonspongy_Ka01=(all['texture']=='nonspongy')&(all['xstal']=='Ka01 crystals: Ka01-26 (mickey), Ka01-18 (Gizmo)')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
ax1.set_title('Ka01 crystals')
x1=P[spongy_Ka01]
x1_sorted=np.sort(x1)
ax1.step(np.concatenate([x1_sorted, x1_sorted[[-1]]]),
         np.arange(x1_sorted.size+1)/x1_sorted.size, color='orange', linewidth=1, label='Spongy')

x1=P[nonspongy_Ka01]
x1_sorted=np.sort(x1)
ax1.step(np.concatenate([x1_sorted, x1_sorted[[-1]]]),
         np.arange(x1_sorted.size+1)/x1_sorted.size, color='blue', linewidth=1, label='nonspongy')
ax1.legend()

x1=comps['Jd'].loc[spongy_Ka01]
x1_sorted=np.sort(x1)
ax2.step(np.concatenate([x1_sorted, x1_sorted[[-1]]]),
         np.arange(x1_sorted.size+1)/x1_sorted.size, color='orange', linewidth=1)

x1=comps['Jd'].loc[nonspongy_Ka01]
x1_sorted=np.sort(x1)
ax2.step(np.concatenate([x1_sorted, x1_sorted[[-1]]]),
         np.arange(x1_sorted.size+1)/x1_sorted.size, color='blue', linewidth=1)
ax1.set_xlabel('Calc P eq32a-32d')
ax2.set_xlabel('Jd')
ax1.set_ylabel('cumulative freq')
```
