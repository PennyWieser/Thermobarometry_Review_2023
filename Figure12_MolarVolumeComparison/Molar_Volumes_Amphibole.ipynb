```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('Molar_Volumes.xlsx', sheet_name='Sheet1')
df_amp=df.loc[0:5, :]
df_cpx=df.loc[6:7, :]
# Reference point is 
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,4))
ax1.plot(df_amp['comp'], df_amp['% delv V'], 'dk', mfc='red')
ax1.plot(df_amp['comp'].iloc[0], df_amp['% delv V'].iloc[0], 'dr', mfc='white')

ax1.plot(df_cpx['comp'], df_cpx['% delv V'], 'ok', mfc='blue')
ax1.plot(df_cpx['comp'].iloc[0], df_cpx['% delv V'].iloc[0], 'ob', mfc='white')

ax2.plot(df_amp['comp'], df_amp['Norm_Solid_Molar_Vol'], 'dk', mfc='red')
ax2.plot(df_amp['comp'].iloc[0], df_amp['Norm_Solid_Molar_Vol'].iloc[0], 'dr', mfc='white')

ax2.plot(df_cpx['comp'], df_cpx['Norm_Solid_Molar_Vol'],  'ok', mfc='blue')
ax2.plot(df_cpx['comp'].iloc[0], df_cpx['Norm_Solid_Molar_Vol'].iloc[0],
         'ob', mfc='white')

for label in ax1.get_xticklabels():
    label.set_rotation(45)
    label.set_ha('right')
    
for label in ax2.get_xticklabels():
    label.set_rotation(45)
    label.set_ha('right')
ax1.set_ylabel('% V of fusion (Liq-Min)')
ax2.set_ylabel('Normalized Molar Volume')

ax3.plot(df_amp['% delv V'], df_amp['Norm_Solid_Molar_Vol'], 'dk', mfc='red')
ax3.plot(df_amp['% delv V'].iloc[0], df_amp['Norm_Solid_Molar_Vol'].iloc[0], 'dr', mfc='white')

ax3.plot(df_cpx['% delv V'], df_cpx['Norm_Solid_Molar_Vol'],  'ok', mfc='blue')
ax3.plot(df_cpx['% delv V'].iloc[0], df_cpx['Norm_Solid_Molar_Vol'].iloc[0],
         'ob', mfc='white')

ax3.set_xlabel('% V of fusion (Liq-Min)')
ax3.set_ylabel('Normalized Molar Volume')

fig.savefig('Amphibole_Volume_Changes.png', dpi=200, bbox_inches='tight')
```
