```python
import numpy as np
import pandas as pd
import Thermobar as pt
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import warnings as w
pd.options.display.max_columns = None
```

```python
df=pd.read_excel('CO2 phase data-101222.xlsx', sheet_name='ALL ISOCHORES')
```

```python
df['Phase'].unique()
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(5,4*0.848))

for un in df['Density (g/mL)'].unique():
    ind=df['Density (g/mL)']==un
    liquid=(df['Phase']=="liquid" ) #|(df['Phase']=="supercritical")
    vapour=df['Phase']=="vapor"
    supercrit=df['Phase']=="supercritical"
    bound=df['Phase']=="liquid and vapor"
    print(un)
    ax1.plot(df['Temperature (C)'].loc[ind&liquid], df['Pressure (MPa)'].loc[ind&liquid], '-b')
    ax1.plot(df['Temperature (C)'].loc[ind&supercrit], df['Pressure (MPa)'].loc[ind&supercrit], '-k')
    ax1.plot(df['Temperature (C)'].loc[ind&vapour], df['Pressure (MPa)'].loc[ind&vapour], '-r')

    ax1.plot(df['Temperature (C)'].loc[ind&bound], df['Pressure (MPa)'].loc[ind&bound], '-c', color='grey', zorder=100)
ax1.plot(31.1, 7.3, '*k', mfc='yellow', ms=15)
    
    #plt.xlim([0, 35])
ax1.plot([25, 25], [0, 10], '-c', linewidth=0.5)
ax1.plot([26, 26], [0, 10], '-c', linewidth=0.5)
ax1.plot([21, 21], [0, 10], '-y', linewidth=0.5)
ax1.plot([22, 22], [0, 10], '-y', linewidth=0.5)

ax1.set_xlim([0, 45])
ax1.set_ylim([0, 10])
fig.savefig('CO2_Phase_Diagram.png', dpi=300, transparent=True)
```
