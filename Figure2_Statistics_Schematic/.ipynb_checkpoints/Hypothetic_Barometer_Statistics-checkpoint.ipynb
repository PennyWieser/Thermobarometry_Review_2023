```python
import Thermobar as pt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
```

```python
x=np.linspace(0, 23, 24)
np.random.seed(5)
# We got a random seed once using this
np.random.rand(24)
```

```python
random=np.array([0.5149222 , 0.93841202, 0.22864655, 0.67714114, 0.59288027,
       0.0100637 , 0.4758262 , 0.70877039, 0.04397543, 0.87952148,
       0.52008142, 0.03066105, 0.22441361, 0.9536757 , 0.58231973,
       0.10747257, 0.2875445 , 0.45670363, 0.02095007, 0.41161551,
       0.48945864, 0.24367788, 0.588639  , 0.75324012])

random2=np.array([0.22199317, 0.87073231, 0.20671916, 0.91861091, 0.48841119,
       0.61174386, 0.76590786, 0.51841799, 0.2968005 , 0.18772123,
       0.08074127, 0.7384403 , 0.44130922, 0.15830987, 0.87993703,
       0.27408646, 0.41423502, 0.29607993, 0.62878791, 0.57983781,
       0.5999292 , 0.26581912, 0.28468588, 0.25358821])
```

```python
# High R2, but Gradient and Intercept wrong
y_highR2=random+x*0.65-1/5+4.65
# Low R2 but perfect match
y_nosys=x+6.9*random-6.9*random2
# Small RMSE but large mean error
y_small_RMSE=random*4.95+x
```

```python
Stats_highR2=pt.calculate_R2_np(x, y_highR2)
print(Stats_highR2['R2'])
print(Stats_highR2['RMSE'])
print(Stats_highR2['Grad'])
print(Stats_highR2['Int'])
print(Stats_highR2['Mean'])
```

```python
Stats_nosys=pt.calculate_R2_np(x, y_nosys)
print(Stats_nosys['R2'])
print(Stats_nosys['RMSE'])
print(Stats_nosys['Grad'])
print(Stats_nosys['Int'])
print(Stats_nosys['Mean'])
```

```python
Stats_smallRMSE=pt.calculate_R2_np(x, y_small_RMSE)
print(Stats_smallRMSE['R2'])
print(Stats_smallRMSE['RMSE'])
print(Stats_smallRMSE['Grad'])
print(Stats_smallRMSE['Int'])
print(Stats_smallRMSE['Mean'])
```
