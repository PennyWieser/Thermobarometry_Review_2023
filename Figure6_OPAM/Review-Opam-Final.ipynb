```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
import Opam

import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
```

```python
# load experimental database
Data = pd.read_excel('OPAM_experimentalDatabase.xlsx')
OPAM = Data.copy()
No_OPAM = Data.copy()

# isolate OPAM saturated experiments
A = (Data['Ol'] == 'Yes') & (Data['Plag'] == 'Yes') & (Data['Cpx'] == 'Yes')
OPAM = OPAM[A]
OPAM['MnO_Liq'][np.isnan(OPAM['MnO_Liq'])] = 0.0
OPAM['Cr2O3_Liq'][np.isnan(OPAM['Cr2O3_Liq'])] = 0.0
OPAM['P2O5_Liq'][np.isnan(OPAM['P2O5_Liq'])] = 0.0
OPAM = OPAM.reset_index(drop = True)
```

```python
## isolate chemical data for calculations
Colls = [col for col in Data.columns if '_Liq' in col]

Liqs = pd.DataFrame()
for C in Colls:
    Liqs[C] = OPAM[C].copy()

Liqs['Sample_ID_Liq'] = np.linspace(0, len(Liqs['SiO2_Liq'])-1, len(Liqs['SiO2_Liq']))

## remove samples without fO2 constraints for V calculations.
Liqs_V = Liqs.copy()
Liqs_V = Liqs_V[OPAM['logfo2'] < 0.0]
Liqs_V = Liqs_V.reset_index(drop = True)
Liqs_V['Sample_ID_Liq'] = np.linspace(0, len(Liqs_V['SiO2_Liq'])-1, len(Liqs_V['SiO2_Liq']))

OPAM_V = OPAM.copy()
OPAM_V = OPAM_V[OPAM['logfo2'] < 0.0]
OPAM_V = OPAM_V.reset_index(drop = True)

No_Liqs = pd.DataFrame()
for C in Colls:
    No_Liqs[C] = No_OPAM[C].copy()

No_Liqs['Sample_ID_Liq'] = np.linspace(0, len(No_Liqs['SiO2_Liq'])-1, len(No_Liqs['SiO2_Liq']))
```

```python
# perform calculations and save results in dictionary
Res_V_all = Opam.pyOpam(liq_comps = No_Liqs, equationP = "Voigt2017", fo2 = "QFM")
Res_V_all_fin = Res_V_all.copy()
Res_V_all_fin =Res_V_all_fin.fillna(0)
V_all_results = pt.calculate_R2(No_OPAM['P_kbar'], Res_V_all['P_kbar_calc'])
V_all_results_test = pt.calculate_R2(No_OPAM['P_kbar'][Res_V_all['Pf'] > 0.8], 
                                     Res_V_all['P_kbar_calc'][Res_V_all['Pf'] > 0.8])
V_all_RMSE = np.sqrt((1/len(No_OPAM['SiO2_Liq']))*np.sum((Res_V_all['P_kbar_calc'] - No_OPAM['P_kbar'])**2))
V_all_RMSE_test = np.sqrt((1/len(No_OPAM['SiO2_Liq'][Res_V_all['Pf'] > 0.8]))*
                     np.sum((Res_V_all['P_kbar_calc'][Res_V_all['Pf'] > 0.8]
                             - No_OPAM['P_kbar'][Res_V_all['Pf'] > 0.8])**2))
V_all_MAE = (1/len(No_OPAM['SiO2_Liq']))*np.sum(Res_V_all['P_kbar_calc'] - No_OPAM['P_kbar'])
V_all_MAE_test = (1/len(No_OPAM['SiO2_Liq'][Res_V_all['Pf'] > 0.8]))*np.sum(Res_V_all['P_kbar_calc'][Res_V_all['Pf'] > 0.8]
                                                                      - No_OPAM['P_kbar'][Res_V_all['Pf'] > 0.8])

Res_V_OPAM = Opam.pyOpam(liq_comps = Liqs_V, equationP = "Voigt2017", fo2 = 10**OPAM_V['logfo2'])
Res_V_OPAM_fin = Res_V_OPAM.copy()
Res_V_OPAM_fin =Res_V_OPAM_fin.fillna(0)
V_OPAM_results = pt.calculate_R2(OPAM_V['P_kbar'], Res_V_OPAM['P_kbar_calc'])
V_OPAM_results_test = pt.calculate_R2(OPAM_V['P_kbar'][Res_V_OPAM['Pf'] > 0.8], 
                                      Res_V_OPAM['P_kbar_calc'][Res_V_OPAM['Pf'] > 0.8])
V_OPAM_RMSE = np.sqrt((1/len(OPAM_V['SiO2_Liq']))*np.sum((Res_V_OPAM['P_kbar_calc'] - OPAM_V['P_kbar'])**2))
V_OPAM_RMSE_test = np.sqrt((1/len(OPAM_V['SiO2_Liq'][Res_V_OPAM['Pf'] > 0.8]))*
                     np.sum((Res_V_OPAM['P_kbar_calc'][Res_V_OPAM['Pf'] > 0.8]
                             - OPAM_V['P_kbar'][Res_V_OPAM['Pf'] > 0.8])**2))
V_OPAM_MAE = (1/len(OPAM_V['SiO2_Liq']))*np.sum(Res_V_OPAM['P_kbar_calc'] - OPAM_V['P_kbar'])
V_OPAM_MAE_test = (1/len(OPAM_V['SiO2_Liq'][Res_V_OPAM['Pf'] > 0.8]))*np.sum(Res_V_OPAM['P_kbar_calc'][Res_V_OPAM['Pf'] > 0.8]
                                                                      - OPAM_V['P_kbar'][Res_V_OPAM['Pf'] > 0.8])

Res_H = Opam.Herzberg2004(Liqs)
Res_H_fin = Res_H.copy()
Res_H_fin = Res_H_fin.fillna(0)
H_results = pt.calculate_R2(OPAM['P_kbar'], Res_H_fin['P_kbar_calc'])
H_MAE = (1/len(OPAM['SiO2_Liq']))*np.sum(Res_H_fin['P_kbar_calc'] - OPAM['P_kbar'])

Res_H_all = Opam.Herzberg2004(No_Liqs)
Res_H_all_fin = Res_H_all.copy()
Res_H_all_fin = Res_H_all_fin.fillna(0)
H_all_results = pt.calculate_R2(No_OPAM['P_kbar'], Res_H_all_fin['P_kbar_calc'])
H_all_MAE = (1/len(No_OPAM['SiO2_Liq']))*np.sum(Res_H_all_fin['P_kbar_calc'] - No_OPAM['P_kbar'])

Res_Y_all = Opam.pyOpam(liq_comps = No_Liqs, equationP = "Yang1996")
Res_Y_all_fin = Res_Y_all.copy()
Res_Y_all_fin =Res_Y_all_fin.fillna(0)
Y_all_results = pt.calculate_R2(No_OPAM['P_kbar'], Res_Y_all['P_kbar_calc'])
Y_all_results_test = pt.calculate_R2(No_OPAM['P_kbar'][Res_Y_all['Pf'] > 0.8], 
                                     Res_Y_all['P_kbar_calc'][Res_Y_all['Pf'] > 0.8])
Y_all_RMSE = np.sqrt((1/len(No_OPAM['SiO2_Liq']))*np.sum((Res_Y_all['P_kbar_calc'] - No_OPAM['P_kbar'])**2))
Y_all_RMSE_test = np.sqrt((1/len(No_OPAM['SiO2_Liq'][Res_Y_all['Pf'] > 0.8]))*
                     np.sum((Res_Y_all['P_kbar_calc'][Res_Y_all['Pf'] > 0.8]
                             - No_OPAM['P_kbar'][Res_Y_all['Pf'] > 0.8])**2))
Y_all_MAE = (1/len(No_OPAM['SiO2_Liq']))*np.sum(Res_Y_all['P_kbar_calc'] - No_OPAM['P_kbar'])
Y_all_MAE_test = (1/len(No_OPAM['SiO2_Liq'][Res_Y_all['Pf'] > 0.8]))*np.sum(Res_Y_all['P_kbar_calc'][Res_Y_all['Pf'] > 0.8]
                                                                      - No_OPAM['P_kbar'][Res_Y_all['Pf'] > 0.8])

Res_Y_OPAM = Opam.pyOpam(liq_comps = Liqs, equationP = "Yang1996")
Res_Y_OPAM_fin = Res_Y_OPAM.copy()
Res_Y_OPAM_fin = Res_Y_OPAM_fin.fillna(0)
Y_OPAM_results = pt.calculate_R2(OPAM['P_kbar'], Res_Y_OPAM['P_kbar_calc'])
Y_OPAM_results_test = pt.calculate_R2(OPAM['P_kbar'][Res_Y_OPAM['Pf'] > 0.8], 
                                      Res_Y_OPAM['P_kbar_calc'][Res_Y_OPAM['Pf'] > 0.8])
Y_OPAM_RMSE = np.sqrt((1/len(OPAM['SiO2_Liq']))*np.sum((Res_Y_OPAM['P_kbar_calc'] - OPAM['P_kbar'])**2))
Y_OPAM_RMSE_test = np.sqrt((1/len(OPAM['SiO2_Liq'][Res_Y_OPAM['Pf'] > 0.8]))*
                     np.sum((Res_Y_OPAM['P_kbar_calc'][Res_Y_OPAM['Pf'] > 0.8]
                             - OPAM['P_kbar'][Res_Y_OPAM['Pf'] > 0.8])**2))
Y_OPAM_MAE = (1/len(OPAM['SiO2_Liq']))*np.sum(Res_Y_OPAM['P_kbar_calc'] - OPAM['P_kbar'])
Y_OPAM_MAE_test = (1/len(OPAM['SiO2_Liq'][Res_Y_OPAM['Pf'] > 0.8]))*np.sum(Res_Y_OPAM['P_kbar_calc'][Res_Y_OPAM['Pf'] > 0.8]
                                                                      - OPAM['P_kbar'][Res_Y_OPAM['Pf'] > 0.8])

Results = {}
Results['Y_OPAM'] = {'Results': Res_Y_OPAM_fin, 'Stats': Y_OPAM_results, 'Stats_test': Y_OPAM_results_test,
                    'MAE': Y_OPAM_MAE, 'MAE_test': Y_OPAM_MAE_test, 'P_kbar_true': OPAM['P_kbar']}

Results['V_OPAM'] = {'Results': Res_V_OPAM_fin, 'Stats': V_OPAM_results, 'Stats_test': V_OPAM_results_test,
                    'MAE': V_OPAM_MAE, 'MAE_test': V_OPAM_MAE_test, 'P_kbar_true': OPAM_V['P_kbar']}

Results['H_OPAM'] = {'Results': Res_H_fin, 'Stats': H_results, 'Stats_test': np.nan,
                    'MAE': H_MAE, 'MAE_test': np.nan, 'P_kbar_true': OPAM['P_kbar']}

Results['Y_all'] = {'Results': Res_Y_all_fin, 'Stats': Y_all_results, 'Stats_test': Y_all_results_test,
                    'MAE': Y_all_MAE, 'MAE_test': Y_all_MAE_test, 'P_kbar_true': Data['P_kbar']}

Results['V_all'] = {'Results': Res_V_all_fin, 'Stats': V_all_results, 'Stats_test': V_all_results_test,
                    'MAE': V_all_MAE, 'MAE_test': V_all_MAE_test, 'P_kbar_true': Data['P_kbar']}

Results['H_all'] = {'Results': Res_H_all_fin, 'Stats': H_all_results, 'Stats_test': np.nan,
                    'MAE': H_all_MAE, 'MAE_test': np.nan, 'P_kbar_true': Data['P_kbar']}
```

```python
# plot results
Outputs = np.array(list(Results.keys())).reshape(2,3)
alpha = 1 # alpha of data that doesn't pass the test
color_true = 'blue' # data that passes the test
color_false = 'red' # data that doesn't pass
true_symbol = 'd' # symbol of data that passes the test
true_size = 8 # size of symbols that pass the test

f, a = plt.subplots(1,3, figsize = (13,4), sharex = True, sharey = True)
f.tight_layout()

a[0].set_xlabel('Experimental P (kbar)')
a[1].set_xlabel('Experimental P (kbar)')
a[2].set_xlabel('Experimental P (kbar)')

a[0].set_ylabel('Calculated P (kbar)')

a[0].set_title('Yang et al. (1996)')
a[1].set_title('Voigt et al. (2017)')
a[2].set_title('Herzberg et al. (2004)')

for i in range(3):
    if type(Results[Outputs[0][i]]['Stats_test']) == dict:
        a[i].plot(Results[Outputs[0,i]]['P_kbar_true'],Results[Outputs[0,i]]['Results']['P_kbar_calc'], 'o',
                     markerfacecolor = color_false, markeredgecolor = 'none', alpha = alpha, label = 'All experiments')
        a[i].plot(Results[Outputs[0,i]]['P_kbar_true'][Results[Outputs[0,i]]['Results']['Pf'] > 0.8],
               Results[Outputs[0,i]]['Results']['P_kbar_calc'][Results[Outputs[0,i]]['Results']['Pf'] > 0.8],
                    markerfacecolor = color_true, markeredgecolor = 'k', markersize = true_size, 
                     marker = true_symbol, linestyle = 'none', label = 'Pass H18 filter')
    else:
        a[i].plot(Results[Outputs[0,i]]['P_kbar_true'],Results[Outputs[0,i]]['Results']['P_kbar_calc'], 'o',
                     markerfacecolor = [0.6,0.6,0.6], markeredgecolor = 'none', alpha = alpha, label = 'All experiments')
    a[i].plot([0,15],[0,15], 'k-')

    if type(Results[Outputs[0][i]]['Stats_test']) == dict:
        a[i].text(7.5, -1, 'R$^{2}$ = ' + Results[Outputs[0,i]]['Stats_test']['R2'], color = color_true)
        a[i].text(7.5, -2, 'RMSE = ' + str(round(float(Results[Outputs[0,i]]['Stats_test']['RMSE']),2)), color = color_true)
        a[i].text(7.5, -3, 'MAE = ' + str(round(Results[Outputs[0,i]]['MAE_test'],2)), color = color_true)
        a[i].text(7.5, -4, 'Grad = ' + str(round(float(Results[Outputs[0,i]]['Stats_test']['Grad']),2)), color = color_true)
        a[i].text(7.5, -5, 'Int = ' + str(round(float(Results[Outputs[0,i]]['Stats_test']['Int']),2)), color = color_true)

    a[i].text(12, -1, 'R$^{2}$ = ' + Results[Outputs[0,i]]['Stats']['R2'], color = [0.6,0.6,0.6])
    a[i].text(12, -2, 'RMSE = ' + str(round(float(Results[Outputs[0,i]]['Stats']['RMSE']),2)), color = [0.6,0.6,0.6])
    a[i].text(12, -3, 'MAE = ' + str(round(Results[Outputs[0,i]]['MAE'],2)), color = [0.6,0.6,0.6])
    a[i].text(12, -4, 'Grad = ' + str(round(float(Results[Outputs[0,i]]['Stats']['Grad']),2)), color = [0.6,0.6,0.6])
    a[i].text(12, -5, 'Int = ' + str(round(float(Results[Outputs[0,i]]['Stats']['Int']),2)), color = [0.6,0.6,0.6])

    a[i].set_xlim([-1, 16])
    a[i].set_ylim([-6, 16])
        
```
